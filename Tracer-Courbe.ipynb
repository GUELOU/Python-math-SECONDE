Représentation Graphique d'une fonction Mathématique

Pour tracer des courbes, la fonction %pylab inline importe les bibliothèques Numpy et Matplotlib.
Il devient alors possible de dessiner des graphiques intégrés au notebook.

```python
#appel de la fonction important les bibliothèques numpy et mathplotlib

%pylab inline 

x =linspace(-pi, pi, 100)# plage de données  de l'axe des abscisses

y=2*x+3 # définition de l'image de x par la fonction 


grid()# tracé de la grille

plot(x,y, label='Courbe Cf') #tracé de l'ensemble des points de coordonnées (x,y)


title("Représentation graphique d'une fonction") # écriture du titre du graphique
xlabel('x') # label sur axe des abscisses
ylabel('y') # label sur axe des ordonnées

legend() # affichage de la légende

savefig('graph1.png') # écriture du graphique dans un fichier image

```
